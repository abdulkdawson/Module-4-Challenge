## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

client_df['subtotal'] = client_df['unit_price']*client_df['qty']

client_df[['unit_price', 'qty', 'subtotal']].head()

```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
client_df['total_weight'] = client_df['unit_weight']*client_df['qty']

client_df['shipping_price'] = client_df['total_weight'].apply(lambda x:x*7 if x>50 else x*10) 

client_df[['qty','unit_weight', 'total_weight','shipping_price']].head()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

client_df['total_price'] = client_df['subtotal']+client_df['shipping_price']

client_df['total_price'] = client_df['total_price']*0.0925 + (client_df['total_price'])

client_df[['total_price', 'subtotal', 'shipping_price']].head()

```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

client_df['cost_per_line'] = client_df['unit_cost']*client_df['qty']

client_df[['unit_cost','qty','shipping_price', 'cost_per_line']].head()
```

```python
# Create a column for the profit of
# each line using line cost and line price

client_df['line_cost'] = client_df['unit_cost']*client_df['qty']

client_df['line_price'] = client_df['unit_price']*client_df['qty']

client_df['line_profit'] = client_df['line_price']-client_df['line_cost']

client_df[['line_price', 'line_cost','line_profit']] .head()




```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

summary_df = client_df.groupby(["client_id"])
summary_df_total = summary_df[['qty', 'shipping_price', 'line_cost', 'line_price', 'line_profit']].sum()
top_5_client_totals = summary_df_total[summary_df_total.index.isin(top_five_clients_list)]
top_5_client_totals = top_5_client_totals.sort_values('line_price', ascending=False).reset_index()
top_5_client_totals
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.


top_5_client_totals_renamed_df = top_5_client_totals.rename(columns={
    'client_id' : 'Client ID',
    'qty': 'Total Units',
    'shipping_price': 'Total Shipping Price',
    'line_price': 'Total Revenue',
    'line_cost' : 'Total Cost',
    'line_profit': 'Total Profit'
})

top_5_client_totals_renamed_df['Total Shipping Price'] = '$' +\
round((top_5_client_totals_renamed_df['Total Shipping Price'].astype(float)/1000000),2).astype(str) + 'M'

top_5_client_totals_renamed_df['Total Cost'] = '$' +\
round((top_5_client_totals_renamed_df['Total Cost'].astype(float)/1000000),2).astype(str) + 'M'

top_5_client_totals_renamed_df['Total Profit'] = '$' +\
round((top_5_client_totals_renamed_df['Total Profit'].astype(float)/1000000),2).astype(str) + 'M'

top_5_client_totals_renamed_df['Total Revenue'] = '$' +\
round((top_5_client_totals_renamed_df['Total Revenue'].astype(float)/1000000),2).astype(str) + 'M'

top_5_client_totals_renamed_df.head()
```

```python
# Sort the updated data by "Total Profit" form highest to lowest


top_5_client_totals_renamed_df = top_5_client_totals_renamed_df.sort_values(by="Total Profit", ascending=False)


top_5_client_totals_renamed_df[['Total Units','Total Shipping Price','Total Revenue','Total Profit']].head()
```
